### Calculation

```
df[:3]
```

```
df["total_ortherrooms"] = df["total_rooms"] - df["total_bedrooms"]
```

```
df[:3][["total_ortherrooms", "total_rooms", "total_bedrooms"]]
```

```
df["median_house_value_vat"] =  df["median_house_value"] + (df["median_house_value"] * 7/100)
```

```
df[["median_house_value_vat", "median_house_value"]][:3]
```

```
df["median_house_value_vat"].describe()
```

```
def transform_price(_price):

    if _price <= 127758:
        return "economics"
    elif _price >= 283550:
        return "luxury"
    else:
        return "market"
```

```
df["price"] = df["median_house_value_vat"].apply(transform_price)
```

```
df[["price", "median_house_value_vat"]].sample(5)
```

```
condition_price = df["price"].str.contains("lux")
```

```
df[condition_price][:5]
```

## Group by Aggregate

```
df["price"].value_counts()
```

```
df.groupby(["price"])[["median_income"]].agg(["min", "mean", "median", "max", "std"])
```

```
df["median_income"].describe()
```

```
def transform_income(_income):

    if _income <= 2.5:
        return "poor"
    elif _income >= 4.7:
        return "rich"
    else:
        return "middle"
```

```
df["income"] = df["median_income"].apply(transform_income)
```

```
df[["income", "median_income"]].sample(5)
```

```
df_group_price_income = df.groupby(["price", "income"])[["median_income"]]\
    .agg(["count", "min", "mean", "median", "max", "std"])
```

```
df_group_price_income.style.background_gradient()
```

## Pivotable

```
df_pivot_price_income = pd.pivot_table(df, 
               values="median_income", 
               index="price", 
               columns="income", 
               aggfunc='mean', 
               fill_value=None)
```

```
df_pivot_price_income
```

```
df_pivot_price_income.style.background_gradient()
```
