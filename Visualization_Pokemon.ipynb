# Visualization

```
import matplotlib.pyplot as plt
import seaborn as sns
%matpltlib inline
```

## BarPlot

```
df_pokemon[:3]
```

## Data Type
1. Numerical Data
2. Categorical Data 
3. Spatial Data 

```
plt.figure(figsize=(15,4))
plt.title("Compare Count of Pokemon Type 1")
sns.countplot(x="Type 1", data=df_pokemon)
```

## BoxPlot

```
df_pokemon.columns
```

```
plt.figure(figsize=(15,4))
sns.boxplot(data=df_pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']])
```

## Scatter Plot

```
plt.figure(figsize=(15,4))
sns.scatterplot(x="HP", y="Attack", data=df_pokemon)
```

```
plt.figure(figsize=(15,4))
sns.lmplot(x="HP", y="Attack", data=df_pokemon, hue="Type 1")
```

## PairPlot

```
plt.figure(figsize=(15,4))
sns.pairplot(data=df_pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', "Type 1"]], hue="Type 1")
```

## SwarmPlot

```
plt.figure(figsize=(15,4))
sns.swarmplot(data=df_pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']])
```

## JointPlot

```
plt.figure(figsize=(15,4))
sns.jointplot(x="HP", y="Attack", data=df_pokemon)
```

## Histogram Plot

```
plt.figure(figsize=(15,4))
sns.distplot(df_pokemon["HP"])
```
